# 3. Opening a csv file from an URL

# 14. Convertors

```python
def rename(name):
    if name == 'Royal Challengers Bangalore':
        return 'RCB'
    elif name == 'Kings XI Punjab':
        return 'KXIP'
    elif name == 'Delhi Daredevils':
        return  'DD'
    elif name == 'Mumbai Indians':
        return 'MI'
    elif name == 'Kolkata Knight Riders':
        return 'KKR'
    elif name == 'Rajasthan Royals':
        return 'RR'
    elif name == 'Deccan Chargers':
        return 'DC'
    elif name == 'Chennai Super Kings':
        return 'CSK'
    elif name == 'Kochi Tuskers Kerala':
        return 'KTK'
    elif name == 'Pune Warriors':
        return 'PW'
    elif name == 'Sunrisers Hyderabad': 
        return 'SH'
    elif name == 'Gujarat Lions':
        return 'GL'
    elif name == 'Rising Pune Supergiants':
        return 'RPS'
    elif name == 'Delhi Capitals':
        return 'DC'
    else: 
        return name

rename('Delhi Capitals')
```

```python
df = pd.read_csv('IPL Matches 2008-2020.csv')
df['team1'].unique()
```

```python
df = pd.read_csv('IPL Matches 2008-2020.csv', converters={'team1':rename})
df['team1'].unique()
```
